```python
# Import dependencies
import pandas as pd
from ast import literal_eval
import warnings
warnings.filterwarnings('ignore')
```

```python
#Create the file and read 
pd.set_option('max_colwidth', 400)
meta_df = pd.read_csv("Resources/movies_metadata.csv")
meta_df.head()
```

```python
meta_df.count()
```

```python
# Check that all adult values are either True or False
meta_df['adult'].value_counts()

# Keep rows in which the "adult" column is False, and then drop the "adult" column
meta_df = meta_df[meta_df['adult'] == 'False'].drop('adult',axis='columns')
```

```python
#Check row count
meta_df.count()
```

```python
#Drop additional columns 
meta_df = meta_df.drop(['belongs_to_collection', 'budget', 'homepage', 'overview', 'popularity', 'poster_path', 'revenue', 'runtime','spoken_languages', 'status', 'tagline', 'video', 'imdb_id', 'title'], axis=1)
meta_df.head(10)
```

```python
#Check row counts 
meta_df.count()
```

```python
#Add movie_id column and delete any duplicates
meta_df['movie_id'] = meta_df['id']
print(len(meta_df))
meta_df.drop_duplicates(subset='movie_id', inplace=True)
print(len(meta_df))
meta_df.head()
```

```python
# Reorder the columns
new_column_order = ["movie_id", "original_title", "release_date", "genres", "production_companies", "production_countries", 
                    "original_language", "vote_average", "vote_count", "id"]

meta_df = meta_df[new_column_order]
meta_df.head()
```

```python
# Convert "release date" column to datetime
meta_df['release_date'] = pd.to_datetime(meta_df['release_date'])
meta_df.head()
```

```python
#Investigate movies before '1950-01-01'
meta_df[(meta_df['release_date'] < '1950-01-01')]
```

```python
#Obtain the index, then use index to drop row, display null values
meta_df[(meta_df['release_date'] < '1950-01-01')].index

meta_df = meta_df.drop(meta_df[(meta_df['release_date'] < '1950-01-01')].index)

meta_df[meta_df['release_date'].isnull()]
```

```python
#Drop the remaining empty rows. 
meta_df= meta_df.dropna(axis=0, how='any')
meta_df.count()
```

```python
#Create genres df
genres_df=meta_df[['movie_id','genres']]
genres_df.head()
```

```python
# Explode list of dictionaries to one dictionary per line
genres_df.genres = genres_df.genres.apply(literal_eval)
genres_df = genres_df.explode('genres', ignore_index=True)

genres_df
```

```python
# Split dictionary into separate columns
genres_df = pd.concat([genres_df, genres_df['genres'].apply(pd.Series)], axis=1)

genres_df
```

```python
#Check for null values
genres_df[genres_df['name'].isnull()]
```

```python
#Drop the remaining empty rows. 
genres_df= genres_df.dropna(axis=0, how='all')
genres_df.count()
```

```python
# Drop unneccessary columns.
clean_genres_df = genres_df.drop(columns=['genres', 'id', 0])

# Rename genres column
clean_genres_df.rename(columns={'name':'genres'}, inplace=True)

clean_genres_df
```

```python
#Save genres.csv to Resources
clean_genres_df.to_csv('Resources/genres.csv', encoding ='utf8', index=False)
```

```python
#Create production_companies df
company_df=meta_df[['movie_id','production_companies']]
company_df.head()
```

```python
# Explode list of dictionaries to one dictionary per line
company_df.production_companies = company_df.production_companies.apply(literal_eval)
company_df = company_df.explode('production_companies', ignore_index=True)

company_df
```

```python
# Split dictionary into separate columns
company_df = pd.concat([company_df, company_df['production_companies'].apply(pd.Series)], axis=1)

company_df
```

```python
#Check for null values
company_df[company_df['name'].isnull()]
```

```python
#Drop the remaining empty rows. 
company_df= company_df.dropna(axis=0, how='all')
company_df.count()
```

```python
# Drop unneccessary columns.
clean_comp_df = company_df.drop(columns=['production_companies', 'id', 0])

# Rename companies column
clean_comp_df.rename(columns={'name':'companies'}, inplace=True)

clean_comp_df
```

```python
#Save companies.csv to Resources
clean_comp_df.to_csv('Resources/companies.csv', encoding ='utf8', index=False)
```

```python
#Create production_countries df
country_df=meta_df[['movie_id','production_countries']]
country_df.head()
```

```python
# Explode list of dictionaries to one dictionary per line
country_df.production_countries = country_df.production_countries.apply(literal_eval)
country_df = country_df.explode('production_countries', ignore_index=True)

country_df
```

```python
# Split dictionary into separate columns
country_df = pd.concat([country_df, country_df['production_countries'].apply(pd.Series)], axis=1)

country_df
```

```python
#Check for null values
country_df[country_df['name'].isnull()]
```

```python
#Drop the remaining empty rows. 
country_df= country_df.dropna(axis=0, how='all')
country_df.count()
```

```python
# Drop unneccessary columns.
clean_country_df = country_df.drop(columns=['production_countries', 0])

# Rename country column
clean_country_df.rename(columns={'name':'countries', 'iso_3166_1':'code'}, inplace=True)

clean_country_df
```

```python
#Save countries.csv to Resources
clean_country_df.to_csv('Resources/countries.csv', encoding ='utf8', index=False)
```

```python
#Back to metadata: now that countries, companies and genres are saved seperately drop additional columns 
clean_meta_df = meta_df.drop(['genres','production_companies', 'production_countries', 'id'], axis=1)

# Rename columns
clean_meta_df.rename(columns={'original_title':'title', 'original_language':'language'}, inplace=True)

clean_meta_df.head(10)
```

```python
#Save clean_metadata.csv to Resources
clean_meta_df.to_csv('Resources/clean_metadata.csv', encoding ='utf8', index=False)
```
