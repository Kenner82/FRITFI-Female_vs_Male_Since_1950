```python
#Create genres df 
genres=meta_df['genres']
genres.count()
```

```python
# Use list comprehension to filter out movies without a genre or genre name: 39,558 with both id and name 
meta_genres = [genre for genre in genres
               if ('id' in genre or 'name' in genre)]

# Obtain the length of the new list
len(meta_genres)
```

```python
# Create a function to clean the genre data
def clean_genre(genre):
    genre = dict(genre) #create a non-destructive copy
    alt_titles = {}
    for key in ['id','name']:
        if key in genre:
            alt_titles[key] = genre[key]
            genre.pop(key)
    if len(alt_titles) > 0:
        genre['alt_titles'] = alt_titles

    return genre
```
